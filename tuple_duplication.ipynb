#**TUPLE DUPLICATION**
The duplicated() method returns a Series with True and False values that describe which rows in the DataFrame are duplicated and not.

Use the subset parameter to specify which columns to include when looking for duplicates. By default all columns are included.

By default, the first occurrence of two or more duplicates will be set to False.

Set the keep parameter to False to also set the first occurrence to True.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'brand': ['Nike', 'Nike', 'Indie', 'Indie', 'Indie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df.duplicated()
```

```python
df.duplicated(keep='last')
```

```python
df.duplicated(keep=False)
```

```python
boolseries=df.duplicated(subset=None,keep=False).value_counts()
print(boolseries)
```

```python
unique=df[~df.duplicated()]
print(unique)
```

Performing the same on a csv file

```python
dataset=pd.read_csv("/content/Ontario_Fuel_Prices_1990_2023.csv")

display(dataset)
```

```python
duplicated=dataset.duplicated(keep=False).value_counts()
print(duplicated)
```

```python
unq=dataset[~dataset.duplicated()]
print("Before")
duplicated=dataset.duplicated(keep=False).value_counts()
print(duplicated)
print("after duplicated removal")
print(unq.duplicated(keep=False).value_counts())
```
